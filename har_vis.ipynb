```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
xtrain=pd.read_table('UCI HAR Dataset/train/X_train.txt',delim_whitespace=True,header=None)
xtest=pd.read_table('UCI HAR Dataset/test/X_test.txt',delim_whitespace=True,header=None)
ytrain=pd.read_table('UCI HAR Dataset/train/y_train.txt',header=None)
ytest=pd.read_table('UCI HAR Dataset/test/y_test.txt',header=None)
X = pd.concat([xtrain, xtest], axis = 0)
y = pd.concat([ytrain, ytest], axis = 0)
```

```python
X.head()
```

```python
y2 = y
y2 = np.asarray(y2)
y2 = pd.DataFrame(y2)
y2[0][367]
```

```python
for i in range(0, len(y2)):
    if(y2[0][i] == 1):
        y2[0][i] = 'Walking'
    elif(y2[0][i] == 2):
        y2[0][i] = 'Walking Upstairs'
    elif(y2[0][i] == 3):
        y2[0][i] = 'Walking Downstairs'
    elif(y2[0][i] == 4):
        y2[0][i] = 'Sitting'
    elif(y2[0][i] == 5):
        y2[0][i] = 'Standing'
    elif(y2[0][i] == 6):
        y2[0][i] = 'Laying'
```

```python
X.shape
```

```python
#PCA algorithm being used for the dimensionality reduction process
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X)
```

```python
X_trans = pca.transform(X)
X_trans = pd.DataFrame(X_trans, columns = ['F1', 'F2'])
```

```python
X_trans.head()
```

```python
X_trans['label'] = y2[0]
X_trans.head()
```

```python
X_trans['label'] = y2[0]
X_trans.tail()
```

```python
import matplotlib
sns.lmplot(x = 'F1', y = 'F2', data = X_trans, hue = 'label',  fit_reg = False, x_jitter = 4.25, y_jitter = 2.4, size = 10)
```

```python
import matplotlib
matplotlib.rcParams.update({'font.size': 20})
```

```python
plt.figure(figsize = (15,10))
sns.countplot(x = 'label', data = X_trans)
plt.tight_layout()
plt.show()
```

```python
#hexagonal joint mapping plot for the data
plt.figure(figsize = (5, 5))
sns.jointplot(x= 'F1', y = 'F2', data = X_trans, kind = 'hex', color = 'red', size = 10 )
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (5, 5))
sns.pairplot(X_trans, hue = 'label', size = 5)
```
